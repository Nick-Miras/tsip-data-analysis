```python
import matplotlib.pyplot as plt
import numpy as np

# Define the linear equation
def linear_eq(x):
    return 2 * x + 1  # Example: y = 2x + 1

# Generate x values
x = np.linspace(-10, 10, 100)
y = linear_eq(x)

# Choose a point to highlight the slope
x1 = 2  # Starting point on x-axis
x2 = 4  # Ending point on x-axis (to compute the run)
y1 = linear_eq(x1)  # Corresponding y value
y2 = linear_eq(x2)  # Corresponding y value

# Plot the linear equation
plt.plot(x, y, label="y = 2x + 1", color="blue")



# Add labels and legend
plt.title("Linear Equation with Rise and Run Highlighted")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black',linewidth=0.5)  # x-axis
plt.axvline(0, color='black',linewidth=0.5)  # y-axis
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for stress and strain
strain = [
    0.00, 2.17, 3.48, 4.96, 6.75, 8.22, 9.77, 11.48, 14.47, 16.15,
    17.54, 18.59, 20.41, 22.63, 24.54, 25.69, 26.24, 27.77, 29.75,
    32.30, 34.17, 35.62, 36.66, 37.14, 37.74
]
stress = [
    0.69, 95.69, 149.11, 212.02, 286.20, 349.02, 415.98, 473.80,
    518.12, 455.82, 500.28, 533.20, 557.65, 568.60, 575.44, 577.15,
    579.13, 580.58, 576.11, 561.39, 537.80, 514.70, 472.03, 444.66, 391.88
]

x1 = strain[3]
x2 = strain[3 + 2]
y1 = stress[3]
y2 = stress[3 + 2]

# Plotting the curve
plt.figure(figsize=(10, 6))
plt.plot(strain, stress, label='Stress-Strain Curve', color='blue')

# Highlight the rise and run
plt.vlines(x=x2, ymin=y1, ymax=y2, colors="gray", linestyles="dashed")  # Vertical line (rise)
plt.hlines(y=y1, xmin=x1, xmax=x2, colors="gray", linestyles="dashed")  # Horizontal line (run)

# Mark the points
plt.text(x1, y1, "\nRun", fontsize=11, verticalalignment='top', horizontalalignment='left', fontweight='bold', color='dimgrey')
plt.text(x2, y2, "  Rise", fontsize=11, verticalalignment='top', horizontalalignment='left', fontweight='bold', color='dimgrey')

# Labels and title
plt.title("Stress vs Strain Curve", fontsize=16)
plt.xlabel("Strain (%)", fontsize=14)
plt.ylabel("Stress (MPa)", fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Yield strength region (approximate based on visual inspection)
yield_index = 8  # Approximate index of yield strength
# plt.annotate(
#     "Yield Strength",
#     xy=(strain[yield_index], stress[yield_index]),
#     xytext=(strain[yield_index]-6, 0),
#     fontsize=12, color='dimgrey'
# )
plt.axvline(strain[yield_index], color='black',linewidth=0.5)  # y-axis

# Ultimate tensile strength (maximum stress point)
uts_index = stress.index(max(stress))
plt.annotate(
    "Strain Hardening",
    xy=(strain[uts_index], stress[uts_index]),
    xytext=(strain[uts_index]-10, 0),
    fontsize=12, color='dimgrey'
)
plt.axvline(strain[uts_index], color='black',linewidth=0.5)  # y-axis

# Failure point (end of the curve)
failure_index = len(strain) - 1
plt.annotate(
    "Necking",
    xy=(strain[failure_index], stress[failure_index]),
    horizontalalignment='left',
    xytext=(strain[failure_index]-6, 0),
    fontsize=12, color='dimgrey'
)
plt.axvline(strain[failure_index], color='black',linewidth=0.5)  # y-axis

# Annotations
# Yield strength region (approximate based on visual inspection)
plt.annotate(
    "Yield Strength",
    xy=(strain[yield_index], stress[yield_index]),
    xytext=(strain[yield_index] + 4, stress[yield_index] - 60),
    arrowprops=dict(facecolor='lightgrey', arrowstyle="-"),
    fontsize=12, fontweight='bold', color='dimgrey'
)

# Ultimate tensile strength (maximum stress point)
plt.annotate(
    "Ultimate Tensile Strength (UTS)",
    xy=(strain[uts_index], stress[uts_index]),
    xytext=(strain[uts_index] - 8, stress[uts_index] - 70),
    arrowprops=dict(facecolor='lightgrey', arrowstyle="-"),
    fontsize=12, fontweight='bold', color='dimgrey'
)

# Failure point (end of the curve)
plt.annotate(
    "Failure Point",
    xy=(strain[failure_index], stress[failure_index]),
    xytext=(strain[failure_index] - 7, stress[failure_index] - 50),
    arrowprops=dict(facecolor='lightgrey', arrowstyle="-"),
    fontsize=12, fontweight='bold', color='dimgrey'
)

plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Annotating the curve with important regions

plt.figure(figsize=(12, 8))
plt.plot(strain, stress, marker='o', label='Stress-Strain Curve', color='b')

# Labels and title
plt.title("Stress vs Strain Curve with Key Areas", fontsize=16)
plt.xlabel("Strain (%)", fontsize=14)
plt.ylabel("Stress (MPa)", fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Annotations
# Yield strength region (approximate based on visual inspection)
yield_index = 8  # Approximate index of yield strength
plt.annotate(
    "Yield Strength",
    xy=(strain[yield_index], stress[yield_index]),
    xytext=(strain[yield_index] + 5, stress[yield_index] - 50),
    arrowprops=dict(facecolor='red', arrowstyle="->"),
    fontsize=12, color='red'
)

# Ultimate tensile strength (maximum stress point)
uts_index = stress.index(max(stress))
plt.annotate(
    "Ultimate Tensile Strength (UTS)",
    xy=(strain[uts_index], stress[uts_index]),
    xytext=(strain[uts_index] - 5, stress[uts_index] + 50),
    arrowprops=dict(facecolor='green', arrowstyle="->"),
    fontsize=12, color='green'
)

# Failure point (end of the curve)
failure_index = len(strain) - 1
plt.annotate(
    "Failure Point",
    xy=(strain[failure_index], stress[failure_index]),
    xytext=(strain[failure_index] - 10, stress[failure_index] - 50),
    arrowprops=dict(facecolor='blue', arrowstyle="->"),
    fontsize=12, color='blue'
)

# Rise and run (slope region)
rise_start, rise_end = 1, 4
run_start, run_end = 0, 3
plt.annotate(
    "Rise",
    xy=(strain[rise_end], stress[rise_end]),
    xytext=(strain[rise_end] + 5, stress[rise_end] + 50),
    arrowprops=dict(facecolor='purple', arrowstyle="->"),
    fontsize=12, color='purple'
)
plt.annotate(
    "Run",
    xy=(strain[run_end], stress[run_end]),
    xytext=(strain[run_end] + 10, stress[run_end] - 50),
    arrowprops=dict(facecolor='orange', arrowstyle="->"),
    fontsize=12, color='orange'
)

plt.legend(fontsize=12)
plt.tight_layout()

# Show the annotated plot
plt.show()
```
